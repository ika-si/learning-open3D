```python
import open3d as o3d
import matplotlib.pyplot as plt
```

```python
# 画像の読み込み
color_raw = o3d.io.read_image("SSII2018_Tutorial_Open3D-master/data/image00100.png")
depth_raw = o3d.io.read_image("SSII2018_Tutorial_Open3D-master/data/depth00100.png")
```

```python
rgbd_image = o3d.geometry.RGBDImage.create_from_color_and_depth(color_raw, depth_raw)
print(rgbd_image)
```

```python
plt.subplot(1,2,1)
plt.title('grayscale image')
plt.imshow(rgbd_image.color, cmap='gray')
plt.subplot(1,2,2)
plt.title('depth image')
plt.imshow(rgbd_image.depth)
```

```python
pcd = o3d.geometry.PointCloud.create_from_rgbd_image(
    rgbd_image, o3d.camera.PinholeCameraIntrinsic(
        o3d.camera.PinholeCameraIntrinsicParameters.PrimeSenseDefault))
```
